```python
import pandas as pd

players_stats = pd.read_excel('EURO_2020_DATA.xlsx', sheet_name='Players stats')
```

```python
import plotly.express as px
```

```python
occurences=players_stats[["PlayerID", "PlayedTime"]]
```

```python
occurences = occurences.groupby("PlayedTime").count()
print(occurences)
```

```python
import numpy as np

weighted_mean = (occurences.index * occurences['PlayerID']).sum() / occurences['PlayerID'].sum()

print(weighted_mean)
```

```python
df = occurences.sort_index()

df['CumulativeWeight'] = df['PlayerID'].cumsum()

total_weight = df['PlayerID'].sum()

weighted_median = df[df['CumulativeWeight'] >= total_weight / 2].index[0]

print(weighted_median)
```

```python
fig=px.histogram(x= occurences.index, y=occurences.values)
fig.show()
```
